```python
import numpy as np
import matplotlib.pylab as plt
from scipy.stats import norm
import math
```

```python
def p_value(x, mu, std):
    t_test = (x-mu)/std
    cdf = norm.sf(t_test, 0)
    p_value = 2*min(cdf,1-cdf)
    return p_value
```

```python
mu = 0
std = 1
pvalue_x2 = []
pvalue_maxarg = []
for i in range(10000):
    x = np.random.normal(loc = mu, scale = std, size = 3)
    pvalue_x2.append(p_value(x[2],0,1))
    pvalue_maxarg.append(p_value(max(x),0,1))
```

```python

plt.hist(pvalue_x2)
plt.show()
```

```python
plt.hist(pvalue_maxarg)
plt.show()
```
